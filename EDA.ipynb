```python
import pandas as pd
import plotly.express as px
```

```python
#se extrae la informacion
autos= pd.read_csv('vehicles_us.csv',sep=',')
```

```python
autos=autos.drop_duplicates().reset_index(drop=True)
autos=autos.dropna(subset=['model_year'])
```

```python

figa = px.histogram(autos, x="odometer") # crear un histograma
figa.show() # crear grfico de dispersin
```

```python
figb = px.scatter(autos, x="odometer", y="price") # crear un grfico de dispersin
figb.show() # crear grfico de dispersin
```
